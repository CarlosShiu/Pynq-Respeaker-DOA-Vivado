```python
#define variable
SOUND_SPEED = 343.2

MIC_DISTANCE_4 = 0.08127
MAX_TDOA_4 = MIC_DISTANCE_4 / float(SOUND_SPEED)

MIC_GROUP_N = 2

tau = [0] * MIC_GROUP_N
theta = [0] * MIC_GROUP_N
```

```python
ol = Overlay("../doa_top.bit")
ol.download()
ac108_init()
```

```python
RESPEAKER_ADDR_1 = 0x43C00000
RESPEAKER_ADDR_2 = 0x41200000
RESPEAKER_RANGE = 0x200
RESPEAKER_OFFSET = 0x00
mmio_1 = MMIO(RESPEAKER_ADDR_1, RESPEAKER_RANGE)
mmio_2 = MMIO(RESPEAKER_ADDR_2, RESPEAKER_RANGE)
```

```python
while(1):
    judge = mmio_2.read(0)
    #print(judge)
    if judge == 1 :
        result = mmio_1.read(0)
        argmax_1 = result % 8
        argmax_2 = result / 8
        if argmax_1 == 0 or argmax_2 == 0 :
            ;
        else:
            shift_1 = argmax_1 - 4
            shift_2 = argmax_2 - 4

            tau[0] = shift_1 / float(16000)
            theta[0] = math.asin(tau[0] / MAX_TDOA_4) * 180 / math.pi

            tau[1] = shift_2 / float(16000)
            theta[1] = math.asin(tau[1] / MAX_TDOA_4) * 180 / math.pi

            if np.abs(theta[0]) < np.abs(theta[1]):
                if theta[1] > 0:
                    best_guess = (theta[0] + 360) % 360
                else:
                    best_guess = (180 - theta[0])
            else:
                if theta[0] < 0:
                    best_guess = (theta[1] + 360) % 360
                else:
                    best_guess = (180 - theta[1])

                best_guess = (best_guess + 90 + 180) % 360

            best_guess = (-best_guess + 120) % 360
            print(best_guess)
    else:
        continue;
        
        
```
